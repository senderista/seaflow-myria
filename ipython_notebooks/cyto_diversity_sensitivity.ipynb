```python
%matplotlib inline

import numpy as np
import pandas as pd
from math import *
import matplotlib.pyplot as plt
```

```python
# import the data
# Need to import all of the diversity estimates for the sensitivity analysis
# a = 32 bins
# b = 30 bins
# c = 28 bins
# d = 26 bins
# e = 24 bins
fileURL = 'https://rest.myria.cs.washington.edu:1776/dataset/user-armbrustlab/program-seaflow/relation-cyto_sens_sds/data?format=csv'
result = pd.read_csv(fileURL)

print(result.columns)
```

```python
# Check lat/lon
x = (result['LON']).values # longitude
y = (result['LAT']).values # latitude

# convert all longitude to longitude east
x[x<0]= 360+x[x<0];
```

```python
# Bin the diversity indices for each case by latitude
# bin the data into bins of 1 degree latitude and calculate average and standard deviation in each bin

za = (result['N0a']).values
zb = (result['N0b']).values
zc = (result['N0c']).values
zd = (result['N0d']).values
ze = (result['N0e']).values

data = y;

bins = np.arange(floor(min(y)),ceil(max(y)),1)
bin_count = np.zeros(len(bins)-1)
digitized = np.digitize(data, bins)
bin_counta = [np.shape(za[digitized == i])[0] for i in range(1, len(bins))]
bin_countb = [np.shape(zb[digitized == i])[0] for i in range(1, len(bins))]
bin_countc = [np.shape(zc[digitized == i])[0] for i in range(1, len(bins))]
bin_countd = [np.shape(zd[digitized == i])[0] for i in range(1, len(bins))]
bin_counte = [np.shape(ze[digitized == i])[0] for i in range(1, len(bins))]

#for i in range(0, len(bins)-1):
 #   bin_count[i] = np.shape(z[digitized == i])[0]

bin_means_a = [za[digitized == i].mean() for i in range(1, len(bins))]
bin_sderr_a = [za[digitized == i].std() for i in range(1, len(bins))]/np.sqrt(bin_counta) # standard error

bin_means_b = [zb[digitized == i].mean() for i in range(1, len(bins))]
bin_sderr_b = [zb[digitized == i].std() for i in range(1, len(bins))]/np.sqrt(bin_countb) # standard error

bin_means_c = [zc[digitized == i].mean() for i in range(1, len(bins))]
bin_sderr_c = [zc[digitized == i].std() for i in range(1, len(bins))]/np.sqrt(bin_countc) # standard error

bin_means_d = [zd[digitized == i].mean() for i in range(1, len(bins))]
bin_sderr_d = [zd[digitized == i].std() for i in range(1, len(bins))]/np.sqrt(bin_countd) # standard error

bin_means_e = [ze[digitized == i].mean() for i in range(1, len(bins))]
bin_sderr_e = [ze[digitized == i].std() for i in range(1, len(bins))]/np.sqrt(bin_counte) # standard error
```

```python
lats = bins[0:len(bins)-1]+0.5

fig2 = plt.figure(2, figsize=(10,8))
plt.axes([0.025, 0.025, 0.95, 0.95])

plt.plot(lats,bin_means_a,'k-', lats,bin_means_b,'-b', lats,bin_means_c,'-r', lats,bin_means_d,'-g', lats,bin_means_e,'-c')
#plt.fill_between(lats,np.asarray(bin_means_a)-np.asarray(bin_sderr_a),np.asarray(bin_means_a)+np.asarray(bin_sderr_a),facecolor='grey')

plt.xlabel('latitude',fontsize=14)
plt.ylabel('N0',fontsize=14)

plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```
