```python
print(type(result['fsc_max'].values))
```

```python
# make a histogram of the min values for each variable
x = ([result['fsc_min_pos'].values], [result['chl_min_pos'].values], [result['pe_min_pos'].values])

fig1 = plb.figure(1, figsize=(15,9))
n, bins, patches = plb.hist(x, 32, normed=0, histtype='bar',
                            color=['r', 'g', 'b'],
                            label=['fsc', 'chl', 'pe'])

plb.title('range of variables below beads')
plb.legend()
plb.figure()
```

```python
# make a histogram of the max values for each variable
y = ([result['fsc_max'].values], [result['chl_max'].values], [result['pe_max'].values])

fig1 = plb.figure(1, figsize=(15,9))
n, bins, patches = plb.hist(y, 32, normed=0, histtype='bar',
                            color=['r', 'g', 'b'],
                            label=['fsc', 'chl', 'pe'])

plb.title('range of variables above beads')
plb.legend()
plb.figure()
```
